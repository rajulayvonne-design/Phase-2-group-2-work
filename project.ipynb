```python
query6 = """ 
SELECT mb.movie_id, r.title, mb.original_title,r.year, mb.runtime_minutes,mb.genres, r.studio, r.domestic_gross, r.foreign_gross, mr.averagerating, mr.numvotes
FROM movie_basics AS mb
JOIN Revenue AS r
ON mb.primary_title = r.title
JOIN movie_ratings AS mr
ON mb.movie_id =  mr.movie_id
"""
pd.read_sql_query(query6, conn)
```

```python
df_data_extracted = pd.read_sql_query(query6, conn)

df_data_extracted.to_csv("Data_for_analysis.csv", index=False)
```

```python
df1 = pd.read_csv("Data_for_analysis.csv")

df1.info()
```

```python
df1["foreign_gross"] = (
    df1["foreign_gross"]
    .astype(str)                 # force everything to string
    .str.replace(",", "")        # remove commas
    .replace("nan", "0")         # replace string nan with 0
    .astype(float)               # convert to numeric
)
```

```python
df1['foreign_gross'].value_counts()
```

```python
df1.info()
```

```python
df1["runtime_minutes"] = df1["runtime_minutes"].fillna(df1["runtime_minutes"].mean())
```

```python
# Check the mode of the genres column
genre_mode = df1['genres'].mode()[0]

# Fill missing values with the mode
df1['genres'].fillna(genre_mode, inplace=True)
```

```python
# Check the mode of the studio column
studio_mode = df1['studio'].mode()[0]

# Fill missing values with the mode
df1['studio'].fillna(genre_mode, inplace=True)
```

```python
df1["domestic_gross"] = (
    df1["domestic_gross"]
    .astype(str)                 # force everything to string
    .str.replace(",", "")        # remove commas
    .replace("nan", "0")         # replace string nan with 0
    .astype(float)               # convert to numeric
)
```
